#### Merge Layers

```python
from tensorflow.keras.layers import Input, Add

in_tensor_1 = Input((1,))
in_tensor_2 = Input((1,))
added = Add()([in_tensor_1, in_tensor_2])
```

#### Create model

```python
from tensorflow.keras import Model

model = Model(inputs=[in_tensor_1, in_tensor_2], outputs=added)

model.summary()
```

### Compile the model

```python
model.compile(optimizer='adam', loss='mean_absolute_error')
```
