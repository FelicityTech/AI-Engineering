##### Simple model for classification and regression

#### Build a simple regressor/classifier

```python
from tensorflow.keras.layers import Input, Dense

input_tensor = Input(shape=(1, ))
output_tensor = Dense(2)(input_tensor)
output_tensor_class = Dense(1, activation='sigmoid')(output_tensor)
```

#### Make a regressor / classifier model

```python
from tensorflow.keras.models import Model
model = Model(inputs=input_tensor, outputs=[output_tensor, output_tensor_class])
model.summary()
```

```python
model.compile(optimizer='adam', loss=['mean_absolute_error', 'binary_crossentropy'])
```

```python
X = games_tourney_train[['seed_diff']]
y_reg = games_tourney_train[['score_1', 'score_2']]
y_class = games_tourney_train[['won']]
model.fit(X, [y_reg, y_class], epochs=100)
```

```python
X = games_tourney_test[['seed_diff']]
y_preg = games_tourney_test[['score_diff']]
y_pclass = games_tourney_test[['won']]
model.evaluate(X, [y_preg, y_pclass], verbose=False)
```
