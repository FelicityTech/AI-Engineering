#### Tweaking your convolution
- Zero Padding in Keras

```python
# No Zero Padding is added: default is 'valid' which means no padding is added. The convolutional layer will only be applied to valid positions of the input, resulting in an output that is smaller than the input.
model.add(Conv2D(10, kernel_size=3, activation='relu', input_shape=(img_rows, img_cols, 1)), padding='valid')
```

```python
# Zero Padding is added: 'same' means that the convolutional layer will be applied to all positions of the input, including those at the borders. To achieve this, Keras will add zero padding around the input, resulting in an output that has the same spatial dimensions as the input.
model.add(Conv2D(10, kernel_size=3, activation='relu', input_shape=(img_rows, img_cols, 1)), padding='same')
```

- Strides in Keras

```python
# Strides: The convolutional layer will be applied to the input with a stride of 1, meaning that the filter will move one pixel at a time across the input. This allows for a more detailed analysis of the input, as the filter will be applied to every possible position of the input.
model.add(Conv2D(10, kernel_size=3, activation='relu', input_shape=(img_rows, img_cols, 1)), strides=1)
```

```python
# Strides: The convolutional layer will be applied to the input with a stride of 2, meaning that the filter will move two pixels at a time across the input. This reduces the spatial dimensions of the output and can help to decrease the computational cost.
model.add(Conv2D(10, kernel_size=3, activation='relu', input_shape=(img_rows, img_cols, 1)), strides=2)
```
