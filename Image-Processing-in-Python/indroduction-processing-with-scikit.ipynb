### Visualizing images in the course

```python
def show_image(image, title='image', cmap_type='gray'):
    import matplotlib.pyplot as plt
    plt.imshow(image, cmap=cmap_type)
    plt.title(title)
    plt.axis('off')
    plt.show()
```

#### What is Histogram of an Image?
- THe Histogram of an image is a graphical representation of the amount of pixes of each intensity value. From 0(pure black) to 255(pure white)
- Color histograms: 
  - Red, Green, Blue

#### Applications of Histograms
 - Analysis
 - Thresholding
 - Brightness and contrast
 - Equalize an image

### Histogram in Matplotlib
- It takes an input array (frequency) and bins as parameters.

```python
# Red color of the image
red = dorcas[:, :, 0]

# Obtain the red Histogram
import matplotlib.pyplot as plt
plt.hist(red.ravel(), bins=256, color='red')
plt.title('Red Histogram')
plt.show()
```

```python
blue = dorcas[:, :, 2]

plt.hist(blue.ravel(), bins=256, color='blue')
plt.title('Blue Histogram')
plt.show()
```

### Getting started with Thresholding
- Partitioning an image into a foreground and background
- By making it black and white
- we do so by setting each pixel to :
  - 255 (white) if pixel > thresh value 
  - 0 (black) of pixel < thresh value

#### Thresholding
Simplest method of image segmentation
 - Isolate objects
   - Object detection
   - Face detection
   - Etc.

Only from grayscale images

```python
# Obtain the optimal threshold value
thresh = 127

# Apply thresholding to the image
binary = red_channel > thresh

# Show the original and thresholded images
show_image(image1, 'Original image')
show_image(binary, 'Thresholded image')
```

#### Inverted thresholding

```python
# Obtain the optimal threshold value
thresh = 127

# Apply the thresholding to the red channel
inverted_binary = red_channel <= thresh

# show the original and thresholded images
show_image(image1, 'Original image')
show_image(inverted_binary, 'Inverted Thresholded image')
```

### Categories
- Global or histogram based: good for uniform backgrounds
- Local or adaptive: for uneven background illumination

### Optimal thresh value
GLobal

Uniform background

```python
# Import the otsu thresholding method
from skimage.filters import threshold_otsu

#  Obtain the optimal threshold value
otsu_thresh = threshold_otsu(image1)

# Apply thresholding to the image
otsu_binary = red_channel > otsu_thresh

# Show the original and thresholded images
show_image(image1, 'Original image')
show_image(otsu_binary, 'Otsu Thresholded image')
```

### Optimal thresh value
Local

Uneven background

```python
# Import the local thresholding method
from skimage.filters import threshold_local

# Set the block size to 35
block_size = 35

#  Obtain the optimal local threshold value
local_thresh = threshold_local(image1, block_size, offset=10)

# Apply local thresholding and obtain the binary image
local_binary = image1 > local_thresh
# Show the original and thresholded images
show_image(image1, 'Original image')
show_image(local_binary, 'Local Thresholded image')
```
